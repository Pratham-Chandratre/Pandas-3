# Working on dictionary

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
```

```python
pd.merge(data1, data2)  # It will extract only those values which are common between both
```

```python
pd.merge(data1, data2, how='left')  # It will call ledt data i.e. data written first data1 complete and right data data2 only common thi
```

```python
pd.merge(data1, data2, how='left')
```

# Conversion by creating function

```python
def con_inr(x):
    return x * 80

df['Fare_inr'] = df['Fare'].apply(con_inr)
```

```python
df
```

```python
def create_flag(x):
    if x < 10:
        return "Cheap"
    elif x >=10 and x<20:
        return "Mid"
    else:
        return "High"
```

```python
df['Payment_class'] = df['Fare'].apply(create_flag)
```
